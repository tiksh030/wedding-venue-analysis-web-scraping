```python
regex = r"(\d+)\s*more"
df1["Amenties_more"] = df1["Amenities"].apply(lambda x: re.findall(regex,x)[0] if re.findall(regex,x) else None)
regex = r"(\d+)-"
df1["Capacity_min"] = df1["Amenities"].apply(lambda x: re.findall(regex,x)[0] if re.findall(regex,x) else None)
regex = r"(\d+)\s"
df1["Capacity_max"] = df1["Amenities"].apply(lambda x: re.findall(regex,x)[0] if re.findall(regex,x) else None)
regex = r"[\d,]+"
df1["Menu_price"] = df1["Price"].apply(lambda x: re.findall(regex,x)[0] if re.findall(regex,x) else None)
regex = r"(\d+)\s*Rooms"
df1["Rooms"] = df1["Amenities"].apply(lambda x: re.findall(regex,x)[0] if re.findall(regex,x) else None)
regex = r"(?:Mumbai|Pune|Bangalore|Chennai|Hyderabad|Lucknow|Delhi)"
df1["Destination"] = df1["Location"].apply(lambda x: re.findall(regex,x)[0] if re.findall(regex,x) else None)
regex = r"(\d+)"
df1["Reviews"] = df1["Review"].apply(lambda x: re.findall(regex,x)[0] if re.findall(regex,x) else None)
df1 = df1.drop("Price",axis=1)
df1 = df1.drop("Amenities",axis=1)
df1 = df1.drop("Review",axis=1)
df1 = df1[["Name","Venue_Categories","Destination","Ratings","Reviews","Menu_price","Capacity_min","Capacity_max","Rooms","Amenties_more"]]
df1["Ratings"]=pd.to_numeric(df1["Ratings"]) 
df1["Reviews"]=pd.to_numeric(df1["Reviews"]) 
df1["Capacity_min"]=pd.to_numeric(df1["Capacity_min"]) 
df1["Capacity_max"]=pd.to_numeric(df1["Capacity_max"]) 
df1["Rooms"] = pd.to_numeric(df1["Rooms"])
df1["Amenties_more"] = pd.to_numeric(df1["Amenties_more"])
df1["Menu_price"] = df1["Menu_price"].str.replace(",","")
df1["Menu_price"] = pd.to_numeric(df1["Menu_price"])
df1.duplicated().sum()
df1 = df1.drop_duplicates()
df1["Rooms"].fillna(df1["Rooms"].median(),inplace=True)
df1["Menu_price"].fillna(df1["Menu_price"].median(),inplace=True)
df1["Capacity_min"].fillna(df1["Capacity_min"].mean(),inplace=True)
df1["Reviews"].fillna(df1["Reviews"].mean(),inplace=True)
df1["Ratings"].fillna(df1["Ratings"].mean(),inplace=True)
df1["Venue_Categories"]=df1["Venue_Categories"].fillna(df1["Venue_Categories"].mode()[0])
df1
```

```python
df1["Menu_price"] = df1["Menu_price"].astype("int")
df1["Capacity_min"] = df1["Capacity_min"].astype("int")
df1["Capacity_max"] = df1["Capacity_max"].astype("int")
df1["Rooms"] = df1["Rooms"].astype("int")
df1["Reviews"] = df1["Reviews"].round(1)
df1["Ratings"] = df1["Ratings"].round(1)
df1.to_csv('cleaning.csv',index = False)
df1
```
