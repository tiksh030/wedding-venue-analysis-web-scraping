```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import time

name_data = []
rating_data = []
price_data = []
review_data = []
location_data = []
type_data = []
amenities_data =[]

pages = 20
print("Progress: ", end="")
cities = ["mumbai","pune","bangalore","chennai","hyderabad","lucknow"]
for city in cities:
    for i in range(1, pages+1):
        URL = f'https://www.wedmegood.com/vendors/{city}/wedding-venues/?page={i}'
        h = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/120.0.0.0 Safari/537.36"
    }
        response = requests.get(URL,headers=h)
        html_code = response.text
        soup = BeautifulSoup(html_code)

        for container in soup.find_all('div', attrs={'class': 'PrimaryVendorCard sc-jTzLTM wFnqD'}):
            # 1. 
            name = container.find('div', attrs={'class': 'frow'})
            name_data.append(name.text if name else np.nan)
    
            # 2. Weight & Flavour
            rating = container.find('span', attrs={'class': 'StarRatingNew fw-500 center rating-new-5 regular'})
            rating_data.append(rating.text if rating else np.nan)
    
            # 3. Price
            price = container.find('div', attrs={'class': 'frow f-wrap'})
            price_data.append(price.text if price else np.nan)
    
            # 4. Previous Price
            review = container.find('span', attrs={'class': 'review-cnt regular nowrap margin-l-5'})
            review_data.append(review.text if review else np.nan)
    
            # 5. Rating
            location = container.find('div', attrs={'class': 'info-icon text-tertiary frow margin-r-10 protected-content jsEvent-onCopy'})
            location_data.append(location.text if location else np.nan)
    
            # 6. Review
            Type = container.find('div', attrs={'class': 'info-icon text-tertiary frow margin-r-10 overflow-hidden'})
            type_data.append(Type.text if Type else np.nan)
    
            # 7. Discount
            amenities = container.find('div', attrs={'class': 'v-center margin-10 protected-content jsEvent-onCopy'})
            amenities_data.append(amenities.text if amenities else np.nan)
    
        
    
        time.sleep(2)
        print(".", end="")

print("\nN U M B E R 's")
print(f"{'Name':<25}: {len(name_data)}")
print(f"{'Review':<25}: {len(review_data)}")
print(f"{'Ratings':<25}: {len(rating_data)}")
print(f"{'Price':<25}: {len(price_data)}")
print(f"{'Location':<25}: {len(location_data)}")
print(f"{'Venue_Categories':<25}: {len(type_data)}")
print(f"{'Amenities':<25}: {len(amenities_data)}")
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'Name': name_data,
    'Venue_Categories': type_data,
    'Location': location_data,
    'Ratings': rating_data,
    'Review': review_data,
    'Price': price_data,
    'Amenities': amenities_data
})

df1.to_csv('wedding.csv',index = False)
df1
```
